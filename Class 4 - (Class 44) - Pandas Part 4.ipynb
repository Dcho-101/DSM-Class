```python
# convert the fare from euro to rupees and make a new column fare_rupees


# using apply function this can be done
```

```python
df['Fare_INR'] = df['Fare'].apply(lambda x: x*80)
```

```python
df.head()
```

```python
# also you can implement this function without using lambda
```

```python
def euro_to_inr(x) :
    return x*80


df['Fare_INR'] = df['Fare'].apply(euro_to_inr)
```

```python
df.head()
```

```python
# changing the name or data or categorize the fare from the dataset in fare values.  
```

```python
def cat_fare(x) :
    if x < 10 :
        return "cheap"
    elif x >= 10 and x < 20 :
        return "mid"
    else :
        return "high"
```

```python
df['cat_fare'] = df['Fare'].apply(cat_fare)
```
