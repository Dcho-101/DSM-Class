# Isolation Forest Anomaly Detection

```python
import pandas as pd
df = pd.read_csv('healthcare.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(contamination=0.4)
clf.fit(df)
prediction= clf.predict(df)
```

```python
prediction
```

```python
import numpy as np
index = np.where(prediction < 0)
index
```

```python
X = df.values
```

```python
index = np.where(prediction < 0)
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.scatter(X[index,0],X[index,1],edgecolors='r')            
```

```python
clf = IsolationForest(contamination=0.1) # most of the scene it will be auto
clf.fit(df)
prediction= clf.predict(df)
```

```python
import numpy as np
index = np.where(prediction < 0)
index
```

```python
X = df.values
```

```python
index = np.where(prediction < 0)
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.scatter(X[index,0],X[index,1],edgecolors='r')   
```
