# Decision Tree Classifier Implementation

```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
# another method to load the dataset 

import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
dataset.target
```

```python
# Independent and Dependent Featues
```

```python
X = df.iloc[:,:-1]
y = dataset.target
```

```python
X,y
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
# for doing decision tree, does not need to do the feature selection 
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## Post Pruning

treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))

tree.plot_tree(treeclassifier,filled=True)
```

```python
## Post Pruning with max depth = 2

treeclassifier = DecisionTreeClassifier(max_depth=2)
treeclassifier.fit(X_train,y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))

tree.plot_tree(treeclassifier,filled=True)
```

```python
# Prediction 

y_pred = treeclassifier.predict(X_test)

```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
```

```python
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

## Decision Tree Pretuning and Hyperparameter Tuning for Huge Data

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
parameter = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best', 'random'],
    'max_depth' : [1,2,3,4,5],
    'max_features' : ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeclassifier = DecisionTreeClassifier()

clf = GridSearchCV(treeclassifier,param_grid=parameter, cv = 5, scoring='accuracy')
```

```python
# training the dataset

clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(X_test)
```

```python
score = accuracy_score(y_pred,y_test)
```

```python
score
```

```python
print(classification_report(y_pred,y_test))
```
