```python
# in this libraru also some datasets are availabe

from bokeh.plotting import figure, output_file, show

from bokeh.sampledata.iris import flowers # importing flowers column from the iris dataset. this dataset also availbale in the bokeh library

output_file('test_bokeh.html') # output will show in the console but we have to name it

p = figure(title = "Test Bokeh of Flower") #giving title

p.xaxis.axis_label = "Petal Length" # naming x axis

p.yaxis.axis_label = "Petal Width" # naming y axis

p.circle(flowers['petal_length'], flowers['petal_width']) #making final circle outcome

show(p)
```

```python
# making different graph

from bokeh.plotting import figure, output_file, show

from bokeh.sampledata.iris import flowers

output_file('test_line_bokeh.html')

p = figure(title = "Line Graph of Flower")

p.xaxis.axis_label = "X axis"

p.yaxis.axis_label = "Y axis"

p.line(flowers['petal_length'], flowers['petal_width'])

show(p)
```

```python
# making another graph

a = [2,3,4,5,6]
b = [3,4,5,6,7]

output_file('line.htmml')
p = figure(title = "Line Plot")
p.line(a,b)
show(p)
```

```python
# scatterplot

a = [2,3,4,5,6]
b = [3,4,5,6,7]

output_file('line.htmml')
p = figure(title = "Line Plot")
p.scatter(a,b, fill_color = 'red') # changing mark colour to red
show(p)
```

```python
a = [2,3,4,5,6]
b = [3,4,5,6,7]

output_file('line.htmml')
p = figure(title = "Line Plot")
p.scatter(a,b, fill_color = 'red', legend_label = 'red points', size = 10)  # for showing the red points name in the graph and changing size
show(p)
```
