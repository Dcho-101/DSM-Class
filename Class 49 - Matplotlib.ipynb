```python
x = np.linspace(1,10, 200)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y, '--r') # for decoration --r. -- stands for the line design and r stands for colour
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.show()
```

```python
plt.plot(x,y, '--b') # for decoration --r. -- stands for the line design and r stands for colour
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.show()
```

```python
# controling the figure and size

plt.figure(figsize=(8,4))
plt.plot(x,y, '--b') # for decoration --r. -- stands for the line design and r stands for colour
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.show()
```

```python
# making gridlines in the background

plt.figure(figsize=(8,4))
plt.plot(x,y, '--b') # for decoration --r. -- stands for the line design and r stands for colour
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.grid()
plt.show()
```
