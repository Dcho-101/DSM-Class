```python
# finding the corelation from the dataset

df.corr()
```

```python
# for plotting the corelation

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True) # annnot for seeing the values
```

```python
df.quality.value_counts().plot(kind='bar')
```

```python
# checking the distribution of fixed acidity

sns.distplot(df['fixed acidity'])
```

```python
# using histplot

sns.histplot(df['fixed acidity'], kde = True)
```

```python
# for combining every plotting

for i in df.columns : 
    sns.histplot(df[i], kde = True)
```

```python
## Categorical plots
```

```python
sns.catplot(x = 'quality', y = 'alcohol', data = df, kind = 'box')
```

```python
# checking relation between the datapoints

sns.scatterplot(x = 'alcohol', y = 'pH', hue = 'quality', data = df)
```
