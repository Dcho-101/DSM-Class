```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('flight_price.xlsx')
```

```python
df.describe()
```

```python
# splitting date of journey to date month and year for model performance 

df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
type(df['Date'][0])
```

```python
type(df['Date'][1])
```

```python
type(df['Date'][2])
```

```python
# so the date month and year columns are still in the string format
```

```python
# changing the columns from string to integer

df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.head(2)
```

```python
df.info()
```

```python
# we dont need the date_of_journey column.

# removing that column

df.drop('Date_of_Journey', axis = 1, inplace = True)
```

```python
df.head()
```

```python
# splitting the Dep_Time into Hour and minutes

df['Hours'] = df['Dep_Time'].str.split(':').str[0]
df['Minutes'] = df['Dep_Time'].str.split(':').str[1]
```

```python
type(df['Date'][0])
```

```python
df.head(2)
```

```python
# changing Hours and Minutes datatype to integer

df['Hours'] = df['Hours'].astype(int)
df['Minutes'] = df['Minutes'].astype(int)
```

```python
type(df['Hours'][0])
```

```python
# splitting the Arrival_Time into only arrival time and ignoring the rest

df['Arrival_Time'].str.split(' ').str[0]
```

```python
# now separarting the time into arival hours and arival minutes

df['Arrival_hours'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_minutes'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
```

```python
df.head(2)
```

```python
# droping the nesessary columns
df.drop('Dep_Time', axis = 1, inplace = True)
df.drop('Arrival_Time', axis = 1, inplace = True)
```

```python
df.head(2)
```

```python
df['Arrival_hours'] = df['Arrival_hours'].astype(int)
df['Arrival_minutes'] = df['Arrival_minutes'].astype(int)
```

```python
df.info()
```

```python
# we dont need route column as the source and destination column avaliable

df.drop('Route', axis = 1, inplace=True)
```

```python
df.head()
```

```python
# splitting duration

df['Duration'].str.split(' ').str[1]
```

```python
df['Duration_Hours'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df['Duration_Minutes'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.head()
```

```python
# change the NaN value with 0 -- Internal assesment

df['Duration_Minutes'] = df['Duration_Minutes'].fillna(0)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop' : 0, '1 stop' : 1, '2 stops' : 2, '3 stops' : 3, '4 stops' : 4, np.nan : 1})
```
