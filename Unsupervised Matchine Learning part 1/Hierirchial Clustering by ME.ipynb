### Doing Feature scale in this scenario for changing the 4 Dimension dataset to 2 Dimension dataset

```python
# Standardization

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(iris_data)
```

```python
X_scaled
```

```python
# Implementing PCA for changing the data from 4D to 2D
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca_scaled = pca.fit_transform(X_scaled)
```

```python
pca
```

```python
pca_scaled
```

```python
pca_scaled.shape
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
# Agglomerative Clustering

## to construct a dendogram

import scipy.cluster.hierarchy as sc

## Plot the dendogram
plt.figure(figsize=(20,7))
plt.title("Dendograms")

## Create Dendogram

sc.dendrogram(sc.linkage(pca_scaled, method='ward'))
plt.title("Dendogram")
plt.xlabel('Sample Index')
plt.ylabel('Eucledian Distance')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=2,affinity='euclidean', linkage='ward')
```

```python
cluster.fit(X_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1], c = cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients = []

# Notice that we need to start at 2 clusters for silhoutte coefficient
for k in range(2,11) :
    agglo = AgglomerativeClustering(n_clusters=k,affinity='euclidean', linkage='ward')
    agglo.fit(X_scaled)
    score = silhouette_score(X_scaled, agglo.labels_)
    silhouette_coefficients.append(score)
```

```python
# plotting 
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel('number of clutters')
plt.ylabel('silhoutte coefficient')
plt.show()
```
